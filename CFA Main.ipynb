```python
# Merge the constructs and their standardized loadings
ave_dict = {}

for construct, items in construct_items.items():
    squared_loadings = loadings_df[
        (loadings_df['rval'] == construct) &
        (loadings_df['lval'].isin(items))
    ]['Est. Std'].astype(float) ** 2
    
    ave = squared_loadings.mean()
    ave_dict[construct] = ave

constructs = list(construct_items.keys())
fl_matrix = pd.DataFrame(index=constructs, columns=constructs)

# Fill Fornell-Larcker matrix
for c1 in constructs:
    for c2 in constructs:
        if c1 == c2:
            fl_matrix.loc[c1, c2] = round(np.sqrt(ave_dict[c1]), 3)
        else:
            # For between-construct correlations, extract from your latent variable correlation matrix
            # If not available, placeholder: assume 0.5
            fl_matrix.loc[c1, c2] = "Need latent correlation matrix"

print("📊 Fornell-Larcker Matrix:")
print(fl_matrix)
```

```python
def compute_htmt(data, construct_items):
    constructs = list(construct_items.keys())
    htmt_matrix = pd.DataFrame(index=constructs, columns=constructs)

    for c1 in constructs:
        for c2 in constructs:
            if c1 == c2:
                htmt_matrix.loc[c1, c2] = 1.0
            else:
                items1 = construct_items[c1]
                items2 = construct_items[c2]
                cor_matrix = data[items1 + items2].corr()

                # Heterotrait-Monotrait correlations
                het_trait = cor_matrix.loc[items1, items2].abs().values.flatten()

                # Mono-trait for construct 1
                mt1 = cor_matrix.loc[items1, items1]
                mask1 = np.triu(np.ones(mt1.shape), k=1).astype(bool)
                mask1_df = pd.DataFrame(mask1, index=mt1.index, columns=mt1.columns)
                mono_trait1 = mt1.where(mask1_df).stack().abs().values

                # Mono-trait for construct 2
                mt2 = cor_matrix.loc[items2, items2]
                mask2 = np.triu(np.ones(mt2.shape), k=1).astype(bool)
                mask2_df = pd.DataFrame(mask2, index=mt2.index, columns=mt2.columns)
                mono_trait2 = mt2.where(mask2_df).stack().abs().values

                # Combine and compute HTMT
                denominator = np.mean(np.concatenate([mono_trait1, mono_trait2]))
                htmt_val = np.mean(het_trait) / denominator if denominator != 0 else 0
                htmt_matrix.loc[c1, c2] = round(htmt_val, 3)

    return htmt_matrix

```
